```python
import numpy as np

# Задана функція
def f(x):
    return 1 / x

# Відрізок [1, 2]
x_values = np.linspace(1, 2, 50)
y_values = f(x_values)

x_values, y_values
```

```python
# Обчислення коефіцієнта a за методом найменших квадратів
n = len(x_values)
x_mean = np.mean(x_values)
y_mean = np.mean(y_values)

#чисельник 
numerator = np.sum((x_values - x_mean) * (y_values - y_mean))
#знаменник
denominator = np.sum((x_values - x_mean)**2)

#коефіцієнт многочлена першого порядку
a = numerator / denominator

# Побудова многочлена
F_x = lambda x: a * x

# Оцінка похибки
error = np.sqrt(np.mean((f(x_values) - F_x(x_values))**2))

numerator, denominator, a, error
```

```python
# Виведення результатів
import matplotlib.pyplot as plt

print(f'Коефіцієнт a: {a}')
print(f'Похибка найкращого середньоквадратичного наближення: {error}')

# Графік
plt.plot(x_values, f(x_values), label='f(x) = 1/x')
plt.plot(x_values, F_x(x_values), label=f'F(x) = {a:.4f}x')
plt.scatter(x_values, f(x_values), color='red', label='Точки')
plt.legend()
plt.title('Многочлен найкращого середньоквадратичного наближення')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
