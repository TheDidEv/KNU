```python
import numpy as np
import math

#evenly spaced 10 point on interval
x_val = np.linspace(0, math.pi, 10)

# Задана функція
ourFunc =  np.sin(x_val) ** 2

x_val, ourFunc
```

```python
# Обчислення кроку h
h = x_val[1] - x_val[0]
# Обчислення чисельної похідної (без першої та остагої)
num_pohid = (ourFunc[2:] - ourFunc[1:-1] + ourFunc[:-2]) / (h ** 2)

x_val = x_val[1:-1] 

num_pohid
```

```python
#дійсне значення поїдн.
td = np.sin(x_val) ** 2

#обчислення похибки
e = np.abs(ourFunc[:-2] - num_pohid)

e
```

```python
#індекс мінімальної похібки
m_e = np.argmin(e)

#знаходимо x та похибк удля нього
x_min_e = x_val[m_e]
min_e_val = e[m_e]

x_min_e, min_e_val
```

```python
import matplotlib.pyplot as plt

print(f"x_val: {x_val}")
print(f"x_min_e: {x_min_e}")

plt.figure(figsize=(10, 10))
plt.plot(x_val, e, label="error", color="red")
plt.axvline(x_min_e,linestyle='--',color='blue',label=f'min error on x={x_min_e:.4f}')
plt.title('title')
plt.xlabel('x')
plt.ylabel('error')
plt.legend()
plt.grid(True)
plt.show()
```
