```python
import numpy as np

# Визначення відрізка
x_data = np.linspace(0, 1, 100)

x_data
```

```python
from scipy.optimize import curve_fit

# Оригінальна функція
def f(x):
    return x**3

# Функція многочлена другого ступеня
def poly2(x, a, b, c):
    return a*x**2 + b*x + c

# Генерація значень функції f(x) на відрізку
y_data = f(x_data)

# Пошук коефіцієнтів многочлена методом найменших квадратів
params, covariance = curve_fit(poly2, x_data, y_data)

params, covariance
```

```python
import matplotlib.pyplot as plt

# Коефіцієнти многочлена
a, b, c = params

# Побудова графіка
plt.plot(x_data, y_data, label='Original function: $x^3$')
plt.plot(x_data, poly2(x_data, a, b, c), label='Best fit quadratic polynomial')
plt.legend()

# Обчислення відхилень
deviations = f(x_data) - poly2(x_data, a, b, c)
#середні відхилення
mean_deviation = np.mean(deviations)
#максимальне відхилення
max_deviation = np.max(np.abs(deviations))

print(f"Mean deviation: {mean_deviation}")
print(f"Max deviation: {max_deviation}")

# Показ графіка
plt.show()
```
