```python
import numpy as np

# Визначення функції
def f(x):
    return 1 / x

# Генерація значень функції на відрізку [1, 2]
x_values = np.linspace(1, 2, 1000)
y_values = f(x_values)
# Метод найменших квадратів для побудови многочлена другого степеня
# Створення матриці для системи рівнянь
A = np.zeros((3, 3))
b = np.zeros(3)

# Обчислення скалярних добутків та побудова системи
for i in range(3):
    for j in range(3):
        A[i, j] = np.sum(x_values**(i+j))
    b[i] = np.sum(y_values * x_values**i)

# Розв'язання системи лінійних рівнянь
coefficients = np.linalg.solve(A, b)

# Побудова многочлена
def poly_second_degree(x, coeffs):
    return coeffs[0] + coeffs[1] * x + coeffs[2] * x**2

# Обчислення значень многочлена на відрізку [0, 1]
y_approximated = poly_second_degree(x_values, coefficients)

# Обчислення відхилення
deviation = np.linalg.norm(y_values - y_approximated, ord=np.inf)
print(f"Відхилення: {deviation}")
```

```python
import matplotlib.pyplot as plt
print(f"Відхилення: {deviation}")
# Побудова графіку
plt.figure(figsize=(8, 6))

# Графік функції f(x) = x^3
plt.plot(x_values, y_values, label='f(x) = x^3', color='blue')

# Графік наближеного многочлена
plt.plot(x_values, y_approximated, label='Апроксимація другого степеня', color='red')

plt.title('Найкраще рівномірне наближення f(x) = x^3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
