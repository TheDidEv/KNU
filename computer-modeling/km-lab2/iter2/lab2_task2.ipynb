```python
import numpy as np

# Визначення відрізка
x_data = np.linspace(-1, 1, 100)  # змінено діапазон на [-1, 1]

x_data
```

```python
from scipy.optimize import curve_fit
# Функція Чебишева другого ступеня на інтервалі [-1, 1]
def chebyshev_poly(x):
    return 2 * x**2 - 1

# Функція многочлена другого ступеня
def poly2(x, a, b, c):
    return a * x**2 + b * x + c

# Генерація значень функції f(x) на відрізку
y_data = chebyshev_poly(x_data)

# Пошук коефіцієнтів многочлена методом найменших квадратів
params, covariance = curve_fit(poly2, (x_data + 1) / 2, y_data)  # застосування формули перетворення

# Коефіцієнти многочлена
a, b, c = params

print(f'a = {a}; b = {b}; c = {c}')
print(f'y_date = {y_data}')
print(f'params = {params}')
print(f'covariance = {covariance}')
```

```python
import matplotlib.pyplot as plt
# Побудова графіка
plt.plot((x_data + 1) / 2, y_data, label='Original Chebyshev polynomial: $2x^2 - 1$')
plt.plot((x_data + 1) / 2, poly2((x_data + 1) / 2, a, b, c), label='Best fit quadratic polynomial')
plt.legend()

# Обчислення відхилень
deviations = chebyshev_poly(x_data) - poly2((x_data + 1) / 2, a, b, c)
mean_deviation = np.mean(deviations)
max_deviation = np.max(np.abs(deviations))

print(f"Mean deviation: {mean_deviation}")
print(f"Max deviation: {max_deviation}")

# Показ графіка
plt.show()
```
