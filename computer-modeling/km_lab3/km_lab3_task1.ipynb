```python
import numpy as np
from scipy.integrate import odeint

# Функція, що повертає значення похідної для рівняння y' = sqrt(x) * y + 1
def model(y, x):
    return np.sqrt(x) * y + 1

# Початкова умова
y0 = 0

# Значення x від 1 до 2 з кроком 0.1
x = np.arange(1, 2.1, 0.1)

# Розв'язання рівняння методом Рунге-Кутта 4-го порядку
y_rk4 = odeint(model, y0, x)

# Вивід результатів
print("Метод Рунге-Кутта 4-го порядку:")
for i in range(len(x)):
    print(f"x = {x[i]:.1f}, y = {y_rk4[i][0]}")
```

```python
# Метод Ейлера-Коші для чисельного розв'язання диференціального рівняння
def euler_method(func, y0, x):
    y = np.zeros(len(x))
    y[0] = y0
    h = x[1] - x[0]
    for i in range(1, len(x)):
        y[i] = y[i - 1] + h * func(y[i - 1], x[i - 1])
    return y

# Розв'язання рівняння методом Ейлера-Коші
y_euler = euler_method(model, y0, x)

# Вивід результатів
print("\nМетод Ейлера-Коші:")
for i in range(len(x)):
    print(f"x = {x[i]:.1f}, y = {y_euler[i]}")
```

```python
import matplotlib.pyplot as plt

print(f'Ейлера-Коші: {y_euler}')
print(f'Рунге-Кутта 4-го порядку: {y_rk4}')

# Графік методу Рунге-Кутта 4-го порядку
plt.figure(figsize=(8, 6))
plt.plot(x, y_rk4, label='Рунге-Кутта 4-го порядку', marker='o')
# Графік методу Ейлера-Коші
plt.plot(x, y_euler, label='Ейлер-Коші', marker='x')
# Додавання маркера початкової точки (1, 0)
plt.scatter(1, 0, color='red', label='Початкова точка (1, 0)')

plt.title('Побудова графіку:')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
