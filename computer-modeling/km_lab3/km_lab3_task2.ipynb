```python
import numpy as np

# Функції для системи диференціальних рівнянь
def f(x, y, z):
    return z

def g(x, y, z):
    return -((np.sqrt(x+ y)) / (4 * x))

# Задані початкові умови
x0 = 1
y0 = 1
z0 = 0.5

# Крок
h = 0.1

# Кінцева точка відрізку
xn = 2

# Масиви для зберігання значень x, y та z
x_vals = [x0]
y_vals = [y0]
z_vals = [z0]

# Обчислення значень за методом Ейлера-Коші
while x0 < xn:
    y_next = y0 + h * f(x0, y0, z0)
    z_next = z0 + h * g(x0, y0, z0)
    x0 += h
    y0 = y_next
    z0 = z_next
    
    x_vals.append(x0)
    y_vals.append(y0)
    z_vals.append(z0)


for i in range(len(x_vals)):
    print(f"x = {x_vals[i]}, y = {y_vals[i]}, z = {z_vals[i]}")
```

```python
import matplotlib.pyplot as plt

# Графік розв'язку
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Побудова графіка")
plt.legend()
plt.show()
```
