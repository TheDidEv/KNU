```python
import numpy as np

# Функція, яку інтегруємо
def f(x):
    return 1 / (1 + x**3)

# Метод Сімпсона
def simpson_method(f, a, b, n):
    h = (b - a) / n
    integral = f(a) + f(b)
    
    for i in range(1, n):
        x = a + i * h
        coefficient = 4 if i % 2 != 0 else 2
        integral += coefficient * f(x)
    
    return (h / 3) * integral

# Обчислення інтегралу з точністю ε
def compute_integral(epsilon):
    a = 2  # Нижня межа інтегрування
    b = 1000  # Загальна верхня межа, можна змінити відповідно до потреби
    n1 = 100  # Початкова кількість підрозділів
    n2 = 2 * n1  # Подвоєна кількість підрозділів
    
    integral1 = simpson_method(f, a, b, n1)
    integral2 = simpson_method(f, a, b, n2)
    
    while abs(integral2 - integral1) >= epsilon:
        integral1 = integral2
        n1 = n2
        n2 = 2 * n1
        integral2 = simpson_method(f, a, b, n2)
    
    return integral2

# Обчислення інтегралу з точністю ε = 0.005
epsilon = 0.005
result = compute_integral(epsilon)
print(f"Значення інтегралу з точністю ε = {epsilon}: {result}")
```

```python
import matplotlib.pyplot as plt
# Побудова графіка функції f(x)
x_vals = np.linspace(2, 1000, 1000)
y_vals = f(x_vals)

plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals, label='f(x) = 1 / (1 + x^3)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Графік функції f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
