```python
import numpy as np

# Функція, яку інтегруємо
def f(x):
    return np.exp(x) / np.sqrt(x)

# Метод трапецій Канторовича
def cantorovich_trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    result = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        result += f(a + i * h)
    result *= h
    return result

# Виклик функції для обчислення інтегралу
epsilon = 1e-10  # додатне мале число
a = epsilon  # нижня межа інтегрування
b = 1  # верхня межа інтегрування
n = 1000  # кількість підінтервалів (можна змінювати для отримання точнішого результату)

approx_integral = cantorovich_trapezoidal_rule(f, a, b, n)
print("Наближене значення інтегралу за методом трапецій Канторовича:", approx_integral)
```

```python
import matplotlib.pyplot as plt
x_values = np.linspace(epsilon, 1, 1000)
y_values = f(x_values)

# Побудова графіка функції
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label=r'$\frac{e^x}{\sqrt{x}}$', color='blue')
plt.fill_between(x_values, y_values, alpha=0.2, color='skyblue')  # Підкреслення області інтегрування
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Графік функції e^x / sqrt(x)')
plt.legend()
plt.grid(True)
plt.show()
```
