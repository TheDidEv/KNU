```python
import numpy as np
import matplotlib.pyplot as plt

# Дані
price = np.array([1, 1.25, 1.57, 1.81, 2.09, 2.45, 2.8, 3.19, 3.58, 3.85, 4.5, 5])
demand = np.array([280, 245, 190, 141, 135, 110, 95, 65, 58, 44, 21, 10])
supply = np.array([5, 20, 51, 89, 120, 153, 180, 201, 215, 228, 240, 248])

# Ринкова рівновага без субсидій
equilibrium_index = np.argmin(np.abs(demand - supply))
equilibrium_price = price[equilibrium_index]
equilibrium_quantity = demand[equilibrium_index]

# Вивід ринкової рівноваги без субсидій
print("Ринкова рівновага без субсидій:")
print(f"Ціна: {equilibrium_price}")
print(f"Кількість: {equilibrium_quantity}")

# Розрахунок нових цін з введенням субсидій
subsidy = 0.5  # Припустимо, що субсидія становить 0.5 одиниці
subsidized_price = price + subsidy

# Нова ринкова рівновага з введенням субсидій
new_equilibrium_index = np.argmin(np.abs(demand - supply))
new_equilibrium_price = subsidized_price[new_equilibrium_index]
new_equilibrium_quantity = demand[new_equilibrium_index]

# Вивід нової ринкової рівноваги з введенням субсидій
print("\nНова ринкова рівновага з введенням субсидій:")
print(f"Ціна: {new_equilibrium_price}")
print(f"Кількість: {new_equilibrium_quantity}")

# Графічне відображення
plt.plot(demand, price, label="Попит")
plt.plot(supply, price, label="Пропозиція")
plt.axhline(y=equilibrium_price, color='r', linestyle='--', label="Ринкова рівновага без субсидій")
plt.axhline(y=new_equilibrium_price, color='g', linestyle='--', label="Нова ринкова рівновага з субсидіями")
plt.xlabel('Кількість')
plt.ylabel('Ціна')
plt.title('Ринкова рівновага')
plt.legend()
plt.show()
```
