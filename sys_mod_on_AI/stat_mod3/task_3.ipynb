```python
import random
import matplotlib.pyplot as plt

num_players = 5  
num_flips = 100      
cost_per_flip = 1  
reward = 5          

def simulate_game(num_flips, rule):
    balance = -num_flips * cost_per_flip 
    last_flip = None
    
    for _ in range(num_flips):
        flip = random.choice(["Орел", "Решка"])
        
        if last_flip == "Орел" and flip == rule:
            balance += reward
        last_flip = flip
    
    return balance

results_rule_1 = []  
results_rule_2 = []  

for _ in range(num_players):
    results_rule_1.append(simulate_game(num_flips, "Решка"))
    results_rule_2.append(simulate_game(num_flips, "Орел"))

plt.figure(figsize=(14, 6))
plt.bar(range(1, num_players + 1), results_rule_1, color='blue', alpha=0.6, label="Комбінація 'Орел-Решка'")
plt.bar(range(1, num_players + 1), results_rule_2, color='orange', alpha=0.6, label="Комбінація 'Орел-Орел'")
plt.axhline(0, color='black', linewidth=0.5)
plt.xlabel("Player")
plt.ylabel("Total gain ($)")
plt.title("Winning players according to different rules of the game")
plt.legend()
plt.show()

results_rule_1, results_rule_2
```
