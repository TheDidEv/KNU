```python
import numpy as np
import matplotlib.pyplot as plt

lambda_rate = 5  

num_events = 1000 
inter_arrival_times = np.random.exponential(1 / lambda_rate, num_events)

event_times = np.cumsum(inter_arrival_times)

plt.figure(figsize=(12, 6))
plt.plot(event_times, range(1, num_events + 1), drawstyle="steps-post")
plt.xlabel("Час")
plt.ylabel("Кількість подій")
plt.title("Реалізація Пуассонівського потоку")
plt.grid()
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(inter_arrival_times, bins=30, density=True, color="skyblue", edgecolor="black", alpha=0.7)
plt.xlabel("Інтервал між подіями")
plt.ylabel("Частота")
plt.title("Гістограма інтервалів між подіями")
plt.show()

n_event = 10  
n_event_times = [np.sum(np.random.exponential(1 / lambda_rate, n_event)) for _ in range(1000)]

plt.figure(figsize=(8, 5))
plt.hist(n_event_times, bins=30, density=True, color="lightgreen", edgecolor="black", alpha=0.7)
plt.xlabel("Час появи n-тої події")
plt.ylabel("Частота")
plt.title(f"Гістограма часу появи {n_event}-ої події")
plt.show()

time_interval = 1 
num_events_in_interval = [np.sum(np.cumsum(np.random.exponential(1 / lambda_rate, 100)) <= time_interval) for _ in range(1000)]

plt.figure(figsize=(8, 5))
plt.hist(num_events_in_interval, bins=range(max(num_events_in_interval) + 1), density=True, color="salmon", edgecolor="black", alpha=0.7)
plt.xlabel("Кількість подій за фіксований час")
plt.ylabel("Частота")
plt.title("Гістограма кількості подій за фіксований час")
plt.show()
```
